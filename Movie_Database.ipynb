### Question 3: On average, does Rotten Tomatoes or IMDb give higher ratings? Take into account the fact that Rotten Tomatoes ratings range from 0-100 and IMDb ratings range from 0-10.

```
# your code here
avg_imdb = data["IMDb"].mean()
avg_tomato = data["Rotten Tomatoes"].mean()

avg_tomato = avg_tomato / 10

print(f"Average IMDb rating: {avg_imdb}")
print(f"Average Rotten Tomatoes rating: {avg_tomato}")

if avg_imdb > avg_tomato:
    print("Higher rating: IMDb")
else:
    print("Higher rating: Rotten Tomatoes")
```

### Question 10: From the **cleaned up** dataset, which two columns are the most positively correlated, and which are the most negatively correlated? Disregard the column with itself.

```
correlation_matrix = df_nona.select_dtypes(include=['number']).corr()

corr_pairs = correlation_matrix.unstack()

sorted_corr_pairs = corr_pairs.sort_values(kind="quicksort")

most_positive = sorted_corr_pairs[sorted_corr_pairs < 1].drop_duplicates().tail(1)
most_negative = sorted_corr_pairs.drop_duplicates().head(1)

print("Most Positively Correlated Pair:")
print(most_positive)
print("\nMost Negatively Correlated Pair:")
print(most_negative)
```
